# **Movies On Streaming Platforms Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mov=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/refs/heads/master/MoviesOnStreamingPlatforms_updated.csv')
mov
```

# **DESCRIPTION OF THE COLUMNS**

* ID: Unique identifier for each movie.
* Title: The title of the movie
* Year: The year the movie was released
* Age: The age rating of the movie (e.g., 18+, 7+, all).
* IMDb: The IMDb rating of the movie
* Rotten Tomatoes: A normalized score (0-100) reflecting the overall percentage of positive reviews from critics
* Netflix: Whether the movie is available on Netflix (1 for Yes, 0 for No).
* Hulu: Whether the movie is available on Hulu (1 for Yes, 0 for No).
* Prime Video: Whether the movie is available on Prime Video (1 for Yes, 0 for No).
* Disney+: Whether the movie is available on Disney+ (1 for Yes, 0 for No).
* Directors: The director(s) of the movie.
* Genres: A list of categories classifying the movie's
style and content.
* Country: The country of origin of the movie.
* Language: The language(s) of the movie.
* Runtime: The duration of the movie in minutes.

# Data Cleaning

```python
mov.isnull()
```

```python
mov.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
mov['Directors'].fillna('Unknown', inplace=True)
```

```python
mov['Genres'].fillna('Unknown', inplace=True)
```

```python
mov['Country'].fillna('Unknown', inplace=True)
```

```python
mov['Language'].fillna('Unknown', inplace=True)
```

```python
mov['IMDb'].fillna(5.902751, inplace=True)
```

```python
mov[mov['Runtime'].isnull()]
mov['Runtime'].fillna(93.413447,inplace=True)
```

```python
pd.unique(mov['Type'])
```

```python
mov.drop('Type', axis=1, inplace=True)
```

# Feature Engineering

```python
mov[mov['Rotten Tomatoes'].isnull()]
```

```python
mov['Rotten Tomatoes'] = mov['Rotten Tomatoes'].astype(str)

mov['Rotten Tomatoes'] = mov['Rotten Tomatoes'].str.replace('%', '', regex=True)

mov['Rotten Tomatoes'] = pd.to_numeric(mov['Rotten Tomatoes'], errors='coerce')

mov['Rotten Tomatoes']

```

```python
mov['Rotten Tomatoes'].fillna(mov['Rotten Tomatoes'].mean(), inplace=True)
```

```python
mov['Age'].unique()
```

```python
mov['Age'] = mov['Age'].replace({'all': '0'})
```

```python
mov['Age'] = mov['Age'].astype(str)
mov['Age'] = mov['Age'].str.extract(r'(\d+)', expand=False)

mov['Age'] = pd.to_numeric(mov['Age'], errors='coerce').astype('Int64')
```

```python
mov['Age'].fillna(13,inplace=True)
```

```python
Platform = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
mov_melted = mov.melt(id_vars='Title', value_vars=Platform, var_name='Platforms', value_name='Availability')
mov_melted
```

```python
Platform=['Netflix','Hulu','Prime Video','Disney+']
mov['count_platforms']=mov[Platform].sum(axis=1)
mov['count_platforms']
```

```python
mov['Genres'].str.split(',').explode().value_counts()
```

```python
mov.drop_duplicates(keep=False,inplace=False)
```

```python
pd.to_datetime('06-dec-2024')
```

```python
mov.sort_values('Year',inplace=False)
```

# Group by and Aggregate Methods

```python
grp_obj=mov.groupby('Genres')
grp_obj.groups
```

```python
grp_obj=mov.groupby(['Directors','Country'])
grp_obj.groups
```

```python
grp_obj=mov.groupby(['Country','Language'])
grp_obj.get_group(('India','Hindi'))
```

```python
gp_obj=mov.groupby(['Country'])
a=gp_obj.get_group('India')
a[['Directors','Language']]
```

```python
grp_obj=mov.groupby('Language')
grp_obj.groups
grp_obj.get_group('Chinese,English')
```

```python
director_ratings = mov.groupby('Directors')[['IMDb', 'Rotten Tomatoes']].mean()
filtered_directors = director_ratings[mov['Directors'].value_counts() >= 5]
filtered_directors
```

```python
grp = mov_melted.groupby('Platforms')
a=grp.get_group('Hulu')
b=grp.get_group('Netflix')
c=pd.concat([a,b])
c = pd.merge(c, mov[['Title', 'IMDb']], on='Title', how='left')
c[['Title','IMDb']]
```

# Data visualization

**Distribution of movies across different age ratings**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=mov,palette='rainbow')
plt.title('Distribution of Movies by Age Rating')
plt.xlabel('Age Rating')
plt.ylabel('Number of Movies')
plt.show()
```

**Distribution of movies by genre (top 10 genres)**

```python
plt.figure(figsize=(12, 6))
top_genres = mov['Genres'].str.split(',').explode().value_counts().head(10).index
sns.countplot(x='Genres', data=mov[mov['Genres'].isin(top_genres)], order=top_genres)
plt.title('Distribution of Movies by Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
```

**Distribution of IMDb ratings for movies in the dataset**

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(x='IMDb', data=mov)
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Density')
plt.grid()
```

**Distribution of Rotten Tomatoes scores for movies on Netflix**

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(x='Age', data=mov[mov['Netflix'] == 1],fill=True)
plt.title('Distribution of Rotten Tomatoes Scores for Movies on Netflix')
plt.xlabel('Rotten Tomatoes Score')
plt.ylabel('Density')
plt.grid()
```

**Movie runtimes distributed**

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Runtime', data=mov, bins=50)
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Frequency')
plt.grid()
```

**Distribution of movie release years (with custom bins)**

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Year', data=mov, bins=range(1900, 2021, 5))  # Bins every 5 years
plt.title('Distribution of Movie Release Years (Decades)')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
```

**Relationship between IMDb ratings and Rotten Tomatoes scores for movies**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='IMDb', y='Rotten Tomatoes', data=mov)
plt.title('IMDb Rating vs. Rotten Tomatoes Score')
plt.xlabel('IMDb Rating')
plt.ylabel('Rotten Tomatoes Score')
```

**Relationship between movie runtime and IMDb rating, colored by age rating**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runtime', y='IMDb', data=mov, hue='Age',palette='rainbow')
plt.title('Movie Runtime vs. IMDb Rating, Colored by Age Rating')
plt.xlabel('Runtime (minutes)')
plt.ylabel('IMDb Rating')
```

**IMDb ratings and Rotten Tomatoes scores vary together for movies**

```python
sns.jointplot(x='IMDb', y='Rotten Tomatoes', data=mov)

plt.show()
```

**Joint distribution of IMDb rating and Rotten Tomatoes score, with a regression line**

```python
sns.jointplot(x='IMDb', y='Age', data=mov, kind='reg')
plt.show()
```

**Average IMDb rating by country (top 10 countries)**

```python
plt.figure(figsize=(12, 6))
top_countries = mov['Country'].str.split(',').explode().value_counts().head(10).index
avg_ratings = mov[mov['Country'].isin(top_countries)].groupby('Country')['IMDb'].mean().sort_values(ascending=False)
sns.barplot(x='Country', y='IMDb', data=avg_ratings.reset_index(inplace=False))
plt.title('Average IMDb Rating by Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Average IMDb Rating')
```

**Total Number of Movies on Each Platform**

```python
plt.figure(figsize=(10, 6))
platform_counts = mov[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()
sns.barplot(x=platform_counts.index, y=platform_counts.values,palette='rainbow')
plt.title('Total Number of Movies on Each Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Movies')
plt.show()
```

**IMDb ratings vary across different age ratings**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='IMDb', data=mov)
plt.title('IMDb Ratings by Age Rating')
plt.xlabel('Age Rating')
plt.ylabel('IMDb Rating')
plt.show()
```

**Relationship between IMDb ratings and Rotten Tomatoes scores change for different age ratings**

```python
sns.lmplot(x='IMDb', y='Rotten Tomatoes', data=mov, hue='Age', height=7)
plt.title('IMDb Rating vs. Rotten Tomatoes Score by Age Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Rotten Tomatoes Score')
plt.show()
```

**Distribution of Movies Across Platforms**

```python
platform_counts = mov[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()

plt.figure(figsize=(8, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Movies Across Platforms')
```

**Distribution of Movies Across Age Ratings**

```python
age_counts = mov['Age'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Movies Across Age Ratings')
```

**Correlation Matrix of Numerical Features**

```python
numerical_features = ['Year', 'IMDb', 'Rotten Tomatoes', 'Runtime']
correlation_matrix = mov[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
```
